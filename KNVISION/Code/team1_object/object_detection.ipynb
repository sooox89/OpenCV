```python
import cv2
import numpy as np
import random
```

```python
# 랜덤 색상 생성 함수
def random_color():
    return tuple(np.random.randint(0, 255, 3).tolist())
```

```python
# 처음 등장하는 객체에 랜덤 색상 할당
def assign_random_color(person_id):
    if person_id not in person_colors:
        person_colors[person_id] = random_color()
```

```python
#원본 좌표에서 투시변환된 좌표 출력함수
def original_to_conv(x, y):
    w, h = 1280, 720 # 원본 영상의 해상도
    #원본 영상의 ROI 4개의 좌표(좌상단, 우상단, 좌하단, 우하단)
    srcQuad = np.array([[445, 154], [938, 315], [2, 555], [495, 716]], np.float32)
    #투시 변환 지정
    dstQuad = np.array([[0, 0], [w-1, 0], [0, h-1], [w-1, h-1]], np.float32)
    srcPoint = np.array([[x, y]], dtype=np.float32)
    # 투시변환 진행
    dstPoint = cv2.perspectiveTransform(srcPoint.reshape(-1, 1, 2), cv2.getPerspectiveTransform(srcQuad, dstQuad))
    #메인함수에서 투시변환 창을 0.4배로 resize하였기 때문에 해당 좌표에서 0.4배 적용하여 반환(rezize 배율 변경시 반드시 바꿔줄 것!)
    return int(dstPoint[0, 0, 0])*0.4, int(dstPoint[0, 0, 1])*0.4
```

```python
#투시변환된 좌표에서 원본좌표 출력함수
def conv_to_original(x, y):
    w, h = 1280, 720
    #원본 영상의 ROI 4개의 좌표(좌상단, 우상단, 좌하단, 우하단)
    srcQuad = np.array([[445, 154], [938, 315], [2, 555], [495, 716]], np.float32)
    dstQuad = np.array([[0, 0], [w-1, 0], [0, h-1], [w-1, h-1]], np.float32)
    #메인함수에서 투시변환 창을 0.4배로 resize하였기 때문에 변환된 좌표에서 0.4를 나누어준다(rezize 배율 변경시 반드시 바꿔줄 것!)
    dstPoint = np.array([[x/0.4, y/0.4]], dtype=np.float32)
    # 투시변환 역으로 진행
    srcPoint = cv2.perspectiveTransform(dstPoint.reshape(-1, 1, 2), cv2.getPerspectiveTransform(dstQuad, srcQuad))
    return int(srcPoint[0, 0, 0]), int(srcPoint[0, 0, 1])
```

```python
#마우스 클릭시 이벤트 발생 함수
def mouse_callback(event, x, y, flags, param):
    # 좌표 텍스트와 위치를 표시해주는 변수들을 전역 변수로 선언
    global mouse_text_ori, mouse_text_trans, last_click
    #좌클릭시 이벤트 발생
    if event == cv2.EVENT_LBUTTONDOWN:
        #좌표 변환 함수 사용하여 원본 좌표를 구한다
        ori_x, ori_y = conv_to_original(x, y)
        #print(f'Transformed coordinates: ({x}, {y})')
        #print(f'Original coordinates: ({ori_x}, {ori_y})')
        mouse_text_ori = f'Ori:({ori_x}, {ori_y})'
        mouse_text_trans = f'Trans:({x}, {y})'
        last_click = (ori_x, ori_y)
```

```python
#화면에 띄울 텍스트 전역 변수 선언
mouse_text_ori = ''
mouse_text_trans = ''
# 선언해야 좌표 코드 종료시 좌표 찍은게 사라진다
last_click = None
        
#화면 경로 지정
video_path = "/Users/kwon/Desktop/yolo_image/friend_group_2.mp4"
src = cv2.VideoCapture(video_path)
w, h = 1280, 720
srcQuad = np.array([[445, 154], [938, 315], [2, 555], [495, 716]], np.float32)
dstQuad = np.array([[0, 0], [w-1, 0], [0, h-1], [w-1, h-1]], np.float32)

# 좌표계 선 개수 설정용
grid_height, grid_width = 50, 50

cv2.namedWindow('Transformed Frame with Grid')
cv2.setMouseCallback('Transformed Frame with Grid', mouse_callback)


# 배경 제거를 위해 객체 생성
fgbg = cv2.createBackgroundSubtractorKNN(detectShadows=False)

# 모폴로지 연산을 위한 커널 설정
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

frame_num = 0

object_colors = {}  # 객체 레이블과 해당 객체의 색상을 저장하는 딕셔너리
prev_boxes = {}  # 이전 프레임에서 추출한 객체의 라운딩 박스 정보를 저장하는 딕셔너리
object_paths = {}  # 객체의 이동 경로를 저장하는 딕셔너리

# 사람별 색상 정보를 담을 딕셔너리
person_colors = {}


# 바운딩 박스 위치 보정을 위한 딕셔너리
last_frame_boxes = {}


while src.isOpened():
    current_boxes = {}
    
    
    ret, frame = src.read()
    if not ret:
        break
    frame_num += 1

    pers = cv2.getPerspectiveTransform(srcQuad, dstQuad)
    dst = cv2.warpPerspective(frame, pers, (w, h))

    # 원본 영상과 동인한 크기의 그리드 생성
    grid_img = np.zeros_like(dst)

    # 수평과 수직선 그리기
    for i in range(0, grid_img.shape[0], grid_height):
        cv2.line(grid_img, (0, i), (grid_img.shape[1], i), (0, 255, 0), thickness=1)

    for j in range(0, grid_img.shape[1], grid_width):
        cv2.line(grid_img, (j, 0), (j, grid_img.shape[0]), (0, 255, 0), thickness=1)

    # 기준 노란선 그리기
    cv2.line(grid_img, (9*grid_width, 0), (9*grid_width, grid_img.shape[0]), (0, 255, 255), thickness=2)

    # 그리드 프레임과 변환된 원본 영상 합치기(dst의 배수를 줄일 수록 원본영상이 투명해짐)
    grid_frame = cv2.addWeighted(dst, 0, grid_img, 1, 0)
    # 그리드 크기 리사이즈
    grid_frame = cv2.resize(grid_frame, None, fx=0.4, fy=0.4)
    
    
    # 라스트 클릭의 위치에 빨간 점 생성
    if last_click is not None:
        cv2.circle(frame, last_click, 5, (0, 0, 255), -1)
        # 빨간 점 옆에 위치 나타내는 원본 좌표 표시
        cv2.putText(frame, mouse_text_ori, (last_click[0], last_click[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)
    # 변환된 그리드 창에도 변환된 좌표 텍스트로 표시
    cv2.putText(grid_frame, mouse_text_trans, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)



    # 배경 제거
    background_extraction_mask = fgbg.apply(frame)

    # 배경만 제거하기 위해 반전
    foreground_mask = cv2.bitwise_not(background_extraction_mask)

    # 모폴로지 연산으로 노이즈 제거
    foreground_mask = cv2.morphologyEx(background_extraction_mask, cv2.MORPH_OPEN, kernel)
    
     # Apply Canny edge detection
    edges = cv2.Canny(foreground_mask, 100, 200)
    dilated_mask = cv2.dilate(edges, None, iterations=4)
    dilated_mask = cv2.erode(dilated_mask, None, iterations=2)


    # 프레임 내의 연결된 요소(사람) 찾기
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(dilated_mask)

    # 각 연결된 요소(사람)에 대해 바운딩 박스 그리기
    for i in range(1, num_labels):  # 배경을 제외하기 위해 1부터 시작
        x, y, wid, hit, area = stats[i]
        
        # 작은 영역 제외
        if area < 2500: 
            continue

        # 객체 레이블을 이용하여 색상 선택
        object_label = i

        if object_label in prev_boxes:
            # 이전 프레임에서 이미 색상을 할당한 객체라면, 이전 색상을 유지
            box_color = object_colors[object_label]
        else:
            # 이전 프레임에서 색상을 할당하지 않은 객체라면, 새로운 색상 할당
            if object_label not in object_colors:
                object_colors[object_label] = random_color()
            box_color = object_colors[object_label]

        # 기존 라운딩 박스 크기 대신에 일정한 크기로 고정된 새로운 바운딩 박스 계산
        fixed_width = 100
        fixed_height = 200
        x_fixed = x + wid // 2 - fixed_width // 2
        y_fixed = y + hit // 2 - fixed_height // 2

        # 프레임 위에 바운딩 박스 그리기
        pt1 = (x_fixed, y_fixed)
        pt2 = (x_fixed + fixed_width, y_fixed + fixed_height)
        cv2.rectangle(frame, pt1, pt2, box_color, 2)

        # 현재 프레임에서 추출한 객체의 라운딩 박스 정보 저장
        current_boxes[object_label] = (x_fixed, y_fixed, fixed_width, fixed_height)

        # FPS 값을 설정하세요. 이 값은 비디오의 실제 FPS에 따라 다릅니다.
        FPS = 10

        # 객체의 이동 경로를 저장
        if object_label in object_paths:
            # 마지막 N개의 위치만 저장
            object_paths[object_label].append((x_fixed + fixed_width // 2, y_fixed + fixed_height // 2))
            if len(object_paths[object_label]) > FPS:  # 1초 동안의 프레임 수보다 많은 경우
                object_paths[object_label].pop(0)  # 가장 오래된 위치를 제거
        else:
            object_paths[object_label] = [(x_fixed + fixed_width // 2, y_fixed + fixed_height // 2)]


    # 이전 프레임에서 추출한 객체의 라운딩 박스 그리기
    for label, prev_box in prev_boxes.items():
        if label in current_boxes:
            x, y, wid, hit = prev_box
            box_color = object_colors[label]
            pt1 = (x, y)
            pt2 = (x + wid, y + hit)
            cv2.rectangle(frame, pt1, pt2, box_color, 2)

    # 이전 프레임에서 추출한 객체의 라운딩 박스 정보 갱신
    prev_boxes = current_boxes

    # 객체의 이동 경로를 점으로 그리기
    for label, path in object_paths.items():
        for point in path:
            cv2.circle(frame, point, 2, object_colors[label], -1)
            # 원본 프레임의 좌표를 변형된 그리드 프레임의 좌표로 변환
            trans_x, trans_y = original_to_conv(*point)
            trans_x = int(trans_x)
            trans_y = int(trans_y)

            # 변형된 그리드 프레임에 점 그리기
            cv2.circle(grid_frame, (trans_x, trans_y), 2, object_colors[label], -1)




    # 원본영상 띄워주기
    cv2.imshow("Raw video", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


    # 좌표 변환된 영상 띄워주기
    cv2.imshow("Transformed Frame with Grid", grid_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

src.release()
cv2.destroyAllWindows()
```
